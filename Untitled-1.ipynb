##### Description: This program uses an artificial recurrent neural network called Long Short Term (LSTM) to predict closing stock price of corpration (Apple Inc.) using the last 60 days stock price.

```python
#import libraries 
import math
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import pandas_datareader as web
plt.style.use('fivethirtyeight')
```

```python
# import yfinance to fetch stock data
import yfinance as yf

# Fetch stock data for Apple Inc. from Yahoo Finance
df = yf.download('AAPL', start='2012-01-01', end='2025-04-17')
df.tail(10)
```

```python
print(df.columns)
```

```python
# get the number of rows and columns in the dataset
df.shape
```

```python
# plot the closing price history
plt.figure(figsize=(18,7))
plt.title('Close Price History')
plt.plot(df[('Close', 'AAPL')], label='Close Price')
plt.ylabel('Closing Price($)')
plt.xlabel('Date', fontsize=15)
plt.grid(True)
```

```python
df.shape
```

```python

#create a new dataframe with only the close column
data = df.filter(items=[('Close', 'AAPL')])  # Correctly filter the column



# then convert it to a numpy array
dataset = data.values

#get the number of rows to train the model on
training_data_len = math.ceil(len(dataset) * 0.8)

```

```python
#Scale the data for a model to perform

Scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = Scaler.fit_transform(dataset)  # Ensure dataset is not empty
scaled_data
```

```python
#Create the training dataset
train_data = scaled_data[0:training_data_len, :]

#Split the data into x_train and y_train data sets
x_train = [] #independent varible that we are going to use to predict the y_train
y_train = [] #dependent variable that we are going to predict

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])


        
```

```python
#convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

#Reshape the data to be 3D [samples, time steps, features]
# x_train = np.reshape(x_train, (2625, 60, 1)) 
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1)) # x_train.shape[0] is the number of samples(number of column), x_train.shape[1] is the number of time steps, and 1 is the number of features (the closing price)
```

```python
#Build the LSTM model
model = Sequential()
#input layer
model.add(LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], 1))) #return_sequences=True means that the output of the LSTM layer will be a sequence of outputs, not just the last output. This is important for the next LSTM layer.
#hidden layer
model.add(LSTM(50, return_sequences=False)) #return_sequences=False means that the output of the LSTM layer will be the last output only.
#dropout layer to prevent overfitting
model.add(Dense(25)) #hidden layer Dense layer with 25 neurons means that the output of the LSTM layer will be passed to a Dense layer with 25 neurons.
#hidden layer
model.add(Dense(1)) #output layer 
```

```python
#compile the model
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
#fit the model to the training data
model.fit(x_train, y_train, batch_size=1, epochs=1)  #batch_size=1 means that the model will be trained on one sample at a time, and epochs=1 means that the model will be trained for one epoch only.
```

```python
#Create the testing dataset
test_data = scaled_data[training_data_len - 60: , : ]  # Use the last 60 days of training data for testing

#create x_test and y_test datasets
x_test = [] #independent variable that we are going to use to predict the y_test
y_test = dataset[training_data_len:, :] #dependent variable that we are going to predict

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0]) # append the last 60 days of training data to the x_test dataset
    
```

```python
#convert the data to a numpy array
x_test = np.array(x_test) #convert the x_test dataset to a numpy array

#Reshape the data to be 3D [samples, time steps, features]
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1)) # x_test.shape[0] is the number of samples(number of column), x_test.shape[1] is the number of time steps, and 1 is the number of features (the closing price)

#Get the models predicted price values
predictions = model.predict(x_test) #predict the closing price using the x_test dataset
predictions = Scaler.inverse_transform(predictions) #inverse transform the predictions to get the actual closing price values

#Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(predictions - y_test)**2) #calculate the RMSE by taking the square root of the mean of the squared differences between the predicted and actual closing price values
print(rmse) #print the RMSE value

```

```python
#plot the data
train = data[:training_data_len] #get the training data
test = data[training_data_len:] #get the testing data
test['predictions'] = predictions #add the predictions to the testing data
plt.figure(figsize=(18,7)) #set the figure size
plt.title('Model') #set the title of the plot
plt.xlabel('Date') #set the x-axis label
plt.ylabel('Close Price USD ($)') #set the y-axis label
plt.plot(train[('Close', 'AAPL')], label='Train') #plot the training data
plt.plot(test[('Close', 'AAPL')], label='Test') #plot the testing data
plt.plot(test['predictions'], label='Predictions') #plot the predictions
plt.legend() #show the legend
plt.show() #show the plot

```
